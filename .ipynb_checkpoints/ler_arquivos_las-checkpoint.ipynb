```python
import numpy as np
import matplotlib.pyplot as plt
import random
#---------------------------------#
import las2

import platform
import os

if platform.system() == 'Windows':
    path = "Dados"
else:
    pass

arquivos = [] # lista contendo o caminho ate os arquivos las

# 

for r, d, f in os.walk(path):
    for file in f:
        if '.LAS' in file:
            arquivos.append(os.path.join(r,file))

print(len(arquivos))
# para o caso de ter todas as funções em um diretório específico

#import sys
#import platform

#if platform.system() == 'Windows':
#    sys.path.insert(0, 'C:\\Users\\mario\\Documents\\doutorado\\funcoes_gerais')
#else:
#    sys.path.insert(0, '/home/uff/Documents/doutorado/funcoes_gerais')

for i in range(len(arquivos)) :print(i,arquivos[i])
```

```python
perfis = arquivos.copy()

names = []
for i in range(8):
    names.append(i)


campo = {} # lista com todos os arquivos las
for j in range(len(names)):
    try:
        
        massape = las2.read(perfis[j])

        names_mp = [a['mnemonic'] for a in massape['curve']]
        units_mp = [a['unit'] for a in massape['curve']]
        data = {} # dicionario de conteudo de curvas dentro do las
        
        for i in range(len(names_mp)):
            data[names_mp[i]] = massape['data'][i]

        campo[names[j]] = data
        
        #for i in range(len(units_mp)):
        #    data[units_mp[i]] = massape['data'][i]
        #campo[names[j]] = data
        #print(j)
    except:
        pass
```
